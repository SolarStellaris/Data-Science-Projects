```python
train_df=train
test_df=test
```

```python
X=train_df.drop('Max drift mm',axis=1)
y=train_df['Max drift mm']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
rf_reg = RandomForestRegressor(random_state=42)

rf_reg.fit(X_train, y_train)

rf_pred = rf_reg.predict(X_test)

rf_mse = mean_squared_error(y_test, rf_pred)

print("Random Forest Regressor MSE: {:.4f}".format(rf_mse))
```

```python
y_pred=rf_reg.predict(test_df)
```

```python
df=pd.read_csv("C:/Users/Windows10/OneDrive - ust.edu.ph/Ust Engineering/Titles & Certifications/Certifications/IBM Data Science/Kaggle/Datasets/test.csv")
df['Max drift mm']=y_pred
df['Index']=test.index
od=df[['Index','Max drift mm']]
```
